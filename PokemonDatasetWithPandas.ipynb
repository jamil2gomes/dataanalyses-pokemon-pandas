```python
def fight(pokemon1,pokemon2):
    p1 = df_pokemon[df_pokemon.NAME == pokemon1]
    p2 = df_pokemon[df_pokemon.NAME == pokemon2]
    
    
    if p1.loc[:,'HP':'SPEED'].sum().sum() > p2.loc[:,'HP':'SPEED'].sum().sum():
        return(f'{pokemon1} X {pokemon2} =  {pokemon1} venceu!!!')
    else:
        return(f'{pokemon1} X {pokemon2} =  {pokemon2} venceu!!!')
    
fight('Dragonite','Mega Charizard X')
```
